```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rets(S, rf, rb):
    return np.vectorize(lambda l: rb*l - 1/(2*S**2) * (rb**2 * l**2 - 2 * rb * rf * l + rf**2))

def sharpe(rb, std):
    return np.vectorize(lambda l: rb / std if l == 1 else (l*rb - 0.5 * (l* std)**2)/(l*std))

x = np.linspace(2, 25, num=100)
y = rets(1.0, 0.02, 0.1)(x)
z = sharpe(0.1, 0.1)(x)
```

```python
fig = plt.figure()
ax = plt.subplot()

ax.plot(x, y, label='Mean Return')
ax.plot(x, z, label='Sharpe Ratio')
ax.legend()
plt.savefig('fig.svg')
```
